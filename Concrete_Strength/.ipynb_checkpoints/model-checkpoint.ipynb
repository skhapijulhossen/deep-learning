# Modeling

```python
import tensorflow as tf
tf.__version__
```

```python
class Model:

    def __init__(self,
                 input_dim=1,
                 epochs: int = 10,
                 learning_rate: float = 0.01):
        self.epochs = epochs
        self.learning_rate = learning_rate
        # weights and bias random initialization
        self.weights = tf.Variable(tf.random.uniform((input_dim, ), seed=42),
                                   name='weights')
        self.bias = tf.Variable(1.0, name='bias')
        # Cost Function
        self.compute_loss = lambda ypred, ytrue: tf.reduce_mean(
            tf.square(ypred - ytrue))
        # forward propagation
        self.forward = lambda x: self.weights * x + self.bias

    #train model
    def train(self, x, ytrue):
        for epoch in range(self.epochs):
            with tf.GradientTape() as tape:
                ypred = self.forward(x)
                loss = self.compute_loss(ypred, ytrue)
                #compute gradient
                gradients = tape.gradient(loss, {
                    'weights': self.weights,
                    'bias': self.bias
                })
                self.weights.assign_sub(gradients['weights'] *
                                        self.learning_rate)
                self.bias.assign_sub(gradients['bias'] * self.learning_rate)
                print(f'Loss: {loss}')
```

```python
m = Model(input_dim=8, epochs=5)
m.train(features, target)
```
