```python
import os
from skimage import io
```

## Scaling

```python
imaget = io.imread('data\\Leaf_smut\\DSC_0293.JPG')
```

```python
from skimage import transform
r = transform.rescale(imaget, 0.1,channel_axis=-1, mode='symmetric')
io.imshow(r)
```

```python
os.mkdir('AugmentedData')
```

```python
diseases = ['Bacterial_leaf_blight', 'Brown_spot', 'Leaf_smut']

for d in os.listdir('data'):
    os.mkdir('AugmentedData\\'+str(d))
    for image in os.listdir(f'data\\{d}'):
        img = io.imread(f'data\\{d}\\{image}')
        for factor in [0.2, 0.5]:
            rescaled_img = transform.rescale(
                img, factor, channel_axis=-1, mode='symmetric')
            io.imsave(f'AugmentedData\\{d}\\{str(factor)+image}', rescaled_img)
```

# Exposure

```python
from skimage import exposure
e = exposure.adjust_gamma(imaget, 1.5)
io.imshow(e)
```
